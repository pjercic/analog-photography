# ⚗️ Interactive Simulator: The Microscopic Darkroom

### Welcome, Chemists!
When we take a picture, the film looks completely blank. The chemical reaction that traps the light is so small that we need a microscope to see it. 

Today, we are going to write code to build a microscopic grid of **Silver Bromide** crystals. We will hit them with photons (light), pour virtual Caffenol (coffee developer) on them, and watch the chemical chain reaction happen!

**Teacher Instructions:** Run the code blocks below to simulate the 3 stages of film development: Exposure, Development, and Fixing.

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import ListedColormap

def simulate_darkroom():
    """Simulates the chemical states of silver halide crystals."""
    # We will look at a tiny 10x10 grid of crystals on the film
    grid_size = 10
    
    # Let's draw a simple shape with light (a plus sign '+')
    light_pattern = np.zeros((grid_size, grid_size))
    light_pattern[2:8, 4:6] = 1 # Vertical bar
    light_pattern[4:6, 2:8] = 1 # Horizontal bar

    fig, axes = plt.subplots(1, 4, figsize=(20, 5))
    
    # Define our chemical colors
    # 0: Unexposed Silver Halide (Pale Yellow)
    # 1: Latent Image Speck (Slightly darker, but normally invisible)
    # 2: Developed Metallic Silver (Solid Black)
    # 3: Fixed / Washed Away (Pure White)
    cmap = ListedColormap(['#FFFACD', '#D3D3D3', '#000000', '#FFFFFF'])
    
    # --- STAGE 1: Unexposed Film ---
    stage1 = np.zeros((grid_size, grid_size)) # All crystals are state 0
    axes[0].imshow(stage1, cmap=cmap, vmin=0, vmax=3)
    axes[0].set_title("1. Fresh Film\n(Silver Bromide Crystals)", fontsize=14)
    axes[0].axis('off')

    # --- STAGE 2: The Shutter Clicks (Latent Image) ---
    # Photons hit the film. The exposed crystals turn to state 1.
    stage2 = np.copy(stage1)
    stage2[light_pattern == 1] = 1
    axes[1].imshow(stage2, cmap=cmap, vmin=0, vmax=3)
    axes[1].set_title("2. The Latent Image\n(A few electrons knocked loose)", fontsize=14)
    axes[1].axis('off')

    # --- STAGE 3: The Caffenol Developer ---
    # The developer searches for State 1 crystals and amplifies them to State 2 (Black)
    stage3 = np.copy(stage2)
    stage3[stage3 == 1] = 2 
    axes[2].imshow(stage3, cmap=cmap, vmin=0, vmax=3)
    axes[2].set_title("3. Developer (Caffenol)\n(Silver is reduced to black metal!)", fontsize=14)
    axes[2].axis('off')

    # --- STAGE 4: The Fixer ---
    # The fixer washes away any remaining State 0 crystals, leaving clear film (State 3)
    stage4 = np.copy(stage3)
    stage4[stage4 == 0] = 3
    axes[3].imshow(stage4, cmap=cmap, vmin=0, vmax=3)
    axes[3].set_title("4. The Fixer\n(Unexposed crystals washed away)", fontsize=14)
    axes[3].axis('off')

    plt.tight_layout()
    plt.show()

# Run the Microscopic Darkroom Simulator!
simulate_darkroom()
```

### 🔍 What just happened? Let's analyze the chemistry:

1. **Fresh Film:** The pale yellow blocks represent unexposed $AgBr$ crystals sitting safely in the dark.
2. **The Latent Image:** The light hit in the shape of a plus sign. In the real world, this change is so small you can't actually see it with your eyes. The crystals just hold a tiny magnetic charge.
3. **The Developer:** The coffee and washing soda go to work! The Caffenol acts as an **electron donor**. It finds the charged crystals and dumps electrons onto them, turning them into solid, black metallic silver ($Ag$). 
4. **The Fixer:** If we turned the lights on during step 3, the yellow background crystals would turn black too, ruining the picture! The Fixer acts like soap, safely dissolving the unexposed yellow crystals and washing them down the drain.

### 🧠 The Negative Magic
Look closely at the final picture (Step 4). The parts of the film that were hit by the bright light are now **black**. The parts of the film that were in the shadows washed away and are now **clear**. 

This is why we call it a **Negative**! The brightness of the real world has been perfectly reversed by chemistry.
