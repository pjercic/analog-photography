# 👁️ Interactive Simulator: The Shape-Shifting Shadows

### Welcome, Artists!
Today we are learning about how our eyes and our brains work together to take photographs. 

A shadow is just an area where light is blocked. But did you know that the *same exact object* can cast completely different shadows depending on where the Sun is in the sky? 

**Teacher Instructions:** Run the code blocks below to show the class how shadows can shape-shift into different letters of the alphabet, and how our brains use "Pareidolia" to see faces in the clouds!

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np

def simulate_shadow_alphabet():
    """Simulates how the angle of light changes the shape of a shadow."""
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))
    
    # Let's imagine we have a simple soup can (a cylinder) sitting on the ground.
    
    # 1. High Noon (Sun is directly overhead)
    axes[0].add_patch(patches.Circle((0.5, 0.5), 0.3, color='black', alpha=0.8))
    axes[0].set_title("1. Light from Directly Above\nThe shadow is a circle (Letter 'O')", fontsize=12)
    axes[0].axis('off')
    axes[0].set_xlim(0, 1)
    axes[0].set_ylim(0, 1)

    # 2. Sunset (Sun is low on the horizon, shining from the side)
    axes[1].add_patch(patches.Rectangle((0.3, 0.2), 0.4, 0.6, color='black', alpha=0.8))
    axes[1].set_title("2. Light from the Side\nThe shadow stretches out (Letter 'I')", fontsize=12)
    axes[1].axis('off')
    axes[1].set_xlim(0, 1)
    axes[1].set_ylim(0, 1)

    # 3. Afternoon (Sun is at an angle)
    axes[2].add_patch(patches.Ellipse((0.5, 0.5), 0.3, 0.8, angle=45, color='black', alpha=0.8))
    axes[2].set_title("3. Light at an Angle\nThe shadow leans over (A slash '/')", fontsize=12)
    axes[2].axis('off')
    axes[2].set_xlim(0, 1)
    axes[2].set_ylim(0, 1)

    plt.suptitle("The Shape-Shifting Soup Can", fontsize=18)
    plt.tight_layout()
    plt.show()

# Run the shadow simulator!
simulate_shadow_alphabet()
```

### 🔍 What did we just learn?
The object never changed—it was always just a soup can. But because the *light* moved, the shadow changed from an **O**, to an **I**, to a **/**! 

This is why photographers have to pay close attention to the time of day. The world looks completely different at 12:00 PM than it does at 5:00 PM.

### The Pareidolia Simulator
Now, let's talk about **Pareidolia**. This is the superpower your brain has to find faces hidden in random shapes.



Let's ask the computer to draw a completely random, bumpy cloud. Then, we will see how quickly your brain turns it into a character!

```python
from scipy.ndimage import gaussian_filter

def pareidolia_simulator():
    """Generates random noise to simulate a cloud, then triggers pareidolia."""
    # 1. Generate a random, bumpy "cloud" using math
    np.random.seed(42) # We use a seed so the random cloud looks the same every time
    noise = np.random.rand(100, 100)
    cloud = gaussian_filter(noise, sigma=5) # Blur the noise so it looks fluffy

    fig, axes = plt.subplots(1, 2, figsize=(12, 6))

    # Show just the random cloud
    axes[0].imshow(cloud, cmap='Blues')
    axes[0].set_title("1. Just a random, lumpy cloud.", fontsize=14)
    axes[0].axis('off')

    # Trigger Pareidolia!
    axes[1].imshow(cloud, cmap='Blues')
    # Add two dots for eyes
    axes[1].add_patch(patches.Circle((35, 40), 4, color='black', alpha=0.8)) 
    axes[1].add_patch(patches.Circle((65, 40), 4, color='black', alpha=0.8)) 
    # Add a curved line for a smile
    axes[1].add_patch(patches.Arc((50, 65), 40, 20, theta1=0, theta2=180, angle=180, color='black', linewidth=4)) 
    
    axes[1].set_title("2. Pareidolia!\nYour brain automatically connects the dots.", fontsize=14)
    axes[1].axis('off')

    plt.tight_layout()
    plt.show()

# Run the Pareidolia simulator!
pareidolia_simulator()
```

### 🎨 The Art Rule: Be Curious
The computer didn't actually draw a face; it just drew a cloud, two dots, and a line. *You* are the one who saw the face! 

When you go outside with your camera (or just your eyes), remember that the world is full of hidden shapes, letters, and characters just waiting for you to notice them.
